```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
### 전처리 과정 함수화 시도

def get_new_ftr_name_df(og_name_df):
    ftr_copy_df = pd.DataFrame(data=og_name_df.groupby(by='column_name').cumcount(), columns=['copy_cnt'])
    ftr_copy_df = ftr_copy_df.reset_index()
    new_ftr_copy_df = pd.merge(og_name_df,ftr_copy_df,how='outer')
    new_ftr_copy_df['column_name'] = new_ftr_copy_df[['column_name','copy_cnt']].apply(lambda x: str(x[0])+'_'+str(x[1]) if int(x[1])>0 else x[0], axis=1)
    ftr_names = ftr_names_df['column_name'].values.tolist()   
    return ftr_names
```

```python
og_name_df = pd.read_csv('./HAPT Data Set/features.txt', sep='\s',
                        header= None,
                        names = ['column_name'])

og_name_df = og_name_df.reset_index()
og_name_df
```

```python
# step 1~2 column_name을 기준으로 groupby(by='column_name')
# cumcount()함수를 통해 중복된 값을 살펴본다.
cum_name_sr = og_name_df.groupby(by='column_name').cumcount()

# step 3 위의 결과를 df로 생성
new_name_df = pd.DataFrame(cum_name_sr, columns=['copy_cnt'])
new_name_df = new_name_df.reset_index()

# step 4 위의 new_name_df와 og_name_df를 merge를 합니다 ^^
# cf) index 컬럼을 기준으로...

ftr_names_df = pd.merge(og_name_df,new_name_df, how='outer')
```
