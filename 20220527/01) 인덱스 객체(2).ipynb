ndarray의 [ ] 연산자와 유사한 기능을 DataFrame에 제공하기 위해서 판다스는 ix[ ]연산자를 개발초기에 제공했다. ix[ ]는 ix[0,'Pclass']와 같이 행위치 지정으로 인덱스값 0, 열 위치 지정으로 칼럼 명인 'Pclass'를 입력해 원하는 위치의 데이터를 추출할 수 있다.

열위치 지정은 약간 독특하게도 칼럼 명뿐만 아니라 ix[0,2]는 인덱스 0인 로우의 세 번째 칼럼(칼럼 위치 인덱싱을 0부터 시작하므로)에 있는 데이터를 의미한다.

전자의 경우에는 ix [ ]의 칼럼 명칭(label)기반 인덱싱이며, 후자의 경우는 칼럼 위치(position)기반 인덱싱이다. 아이러니 하게도 이 두가지 방식을 모두 제공하게 되면서 코드가 혼돈을 주거나 가독성이 떨어지면서 ix[ ]는 현재 판다스에서 사라지게(deprecated)되었다고 한다.
(개인적으로는 사라지게 될 정도의 큰 문제는 아니라고 생각한다.)

### 명칭 기반 인덱싱과 위치 기반 인덱싱의 구분

ix [ ]뿐만 아니라 iloc[ ], loc[ ]의 차이를 설명하기 위한 용어 정리를 하겠다. 명칭(Label)기반 인덱싱은 칼럼의 명칭을 기반으로 위치를 지정하는 방식이다. '칼럼 명'같이 명칭으로 열 위치를 지정하는 방식이다. 위치(Position) 기반 인덱싱은 0을 출발점으로 하는 가로축, 세로축 좌표 기반의 행과 열 위치를 기반으로 데이터를 지정한다.

그렇다면 DataFrame의 인덱스값은 명칭 기반 인덱싱일까? 위치 기반 인덱싱일까?

* 만약 data_df의 인덱스의 값이 문자열이 아닌 숫자 값(0,1,2...)라면 data_df.ix[0,:]의 행 위치 값 0은 인덱스값이 0인지, 첫번째 행 위치인지 헷갈리게 된다.


* 다음의 예제로 확인해보자. data_df를 reset_index( )를 수행해 이번에는 기존 인덱스를 칼럼화하고 0부터 시작하는 새로운 인덱스를 추가하겠다. 그리고 이 새로운 인덱스를 다시 1부터 시작하는 인덱스로 바꿔보자.

```python
#data_df를 reset_index()로 새로운 숫자형 인덱스를 생성
data_df_reset = data_df.reset_index()
data_df_reset = data_df_reset.rename(columns={'index':'old_index'})
data_df_reset
```

```python
data_df_reset.index=data_df_reset.index+1
data_df_reset
```

이제 인덱스 값이 1부터 시작하는 integer 형입니다. ix[0,1]이 어떤 값을 반환하는지 확인해보자. 

```python
data_df_reset.ix[0,1]
```

```python
data_df_reset.ix[1,1]
```

결과적으로 DataFrame의 인덱스값은 명칭 기반 인덱싱이라고 간주해야 한다. ix[ ] 의 경우 행과 열 위치에 명칭과 위치 기반 인덱싱 모두를 허용한다. 일반적으로 행 위치에 적용되는 인덱스값과 위치 기반 인덱싱이 integer형일 때 코드 작성에 혼선을 초래할 우려가 있다(물론 명칭기반과 위치기반의 데이터 형이 서로 같을 경우 명칭 기반이 우선한다고 이해하면 큰 혼선이 없을 수도 있다.) 

* 그렇기 때문에 새롭게 명칭 기반 인덱싱 연산자인 loc [ ]와 위치 기반 인덱싱인 iloc[ ]연산자를 도입해 이들의 구분을 명확히 하게 되었다.

### DataFrame iloc[ ] 연산자

iloc [ ]는 위 기반 인덱싱만 허용하기 때문에 행과 열 값으로 integer 또는 integer형의 슬라이싱, 팬시 리스트 값을 입력해줘야 한다.

    CF) 불린 인덱싱은 조건을 기술하므로 이에 제약 받지 않음.

```python
data_df.iloc[0,0]
```

iloc[ ]에 위치 인덱싱이 아닌 명칭을 입력하면 오류가 발생한다.

```python
 data_df.iloc[0,'name']
```

```python
# 행 위치에 입력해도 오류가 발생한다. (행위치에...)
data_df.iloc['one',0]
```

```python
data_df_reset.iloc[0,1]
```

행, 열에 integer 값으로 위치 기반 인덱싱을 나타내야 한다는 점을 제외하면 iloc [ ] 는 ix[ ]와 동일하게 슬라이싱, 팬시 및 불린 인덱싱을 제공한다. 

loc [ ]는 명칭 기반으로 데이터를 추출한다. 따라서 행 위치에는 DataFrame index값을, 그리고 열 위치에는 칼럼 명을 입력해준다. 인덱스값이 'one'의 칼럼 명이 'name'인 데이터를 추출해보자.

```python
data_df.loc['one','Name']
```

```python
data_df_reset.loc[1,'Name']
```

```python
#다음 코든느 오류를 발생합니다 why? 인덱스 값이 0인 행이 없음
data_df_reset.loc[0,'Name']
```

일반적으로 슬라이싱을 '시작값: 종료값'과 같이 지정하면 시작 값 ~ 종료 값 -1까지의 범위를 의미한다. 즉, 0:3이면 0부터 2까지 (0,1,2)를 의미한다. 



    * loc [ ]에 슬라이싱 기호를 적용하면 종료 값-1이 아니라 종료 값까지 포함하는 것을 의미한다.



이는 명칭기반 인덱싱의 특성 때문이다.명칭은 숫자형이 아닐 수 있기 때문에 -1을 할 수가 없다.

```python
data_df
```

```python
print('명칭 기반 ix slicing\n', data_df.ix['one':'Two', 'Name'],'\n')
print('위치 기반 iloc slicing\n', data_df.iloc[0:1, 0],'\n')
print('명칭 기반 loc slicing\n', data_df.loc['one':'Two', 'Name'],'\n')
```

'one':'two'로 ix[ ]를 명칭 기반으로 적용한 경우와 loc[ ]를 적용한 경우에는 2개의 행을 반환했지만, 위치 기반 iloc[0:1]은 0번째 행 위치에 해당하는 1개의 행을 반환했다. 어쩔 수 없이 이러한 부분이 명칭 기반 인덱싱에서 헷갈리는 부분이다.

* 특히 DataFrame의 인덱스가 정수형이면 loc [ ]를 이용시 조심해야 한다.

가령 다음과 같이 loc[1:2, 'Name']에서 1:2는 DataFrame의 인덱스 값에 따른 명칭 기반 인덱싱이므로 인덱스가 1,2인 2개의 데이터를 반환한다. 

```python
data_df_reset
```

```python
print(data_df_reset.loc[1:2, 'Name'])
```

아마  ix, iloc, loc의 이러한 차이를 접하면 상당히 헷갈릴 것이다. 여기서 잠깐 ix, iloc, loc의 문제점과 주의점을 살펴보자.

1) 가장 중요한 것은 명칭 기반 인덱싱과 위치 기반 인덱싱의 차이를 이해하는 것이다. DataFrame의 인덱스나 칼럼명으로 데이터에 접근하는 것은 명칭기반 인덱싱이다. 0부터 시작하는 행,열의 위치 좌표에만 의존하는 것이 위치 기반 인덱싱이다.

2) ix[ ]는 명칭 기반 인덱싱과 위치 기반 인덱싱을 모두 적용할 수 있다. DataFrametaFrame의 인덱스가 숫자 형일 경우 행 위치에 오는 숫자는 위치 기반 인덱싱이 아닌 명칭 기반 인덱싱의 DataFrame 인덱스를 말한다.

3) iloc[ ]는 위치 기반 인덱싱만 가능하다. 따라서 행과 열 위치 행과 열 위치 값으로 정수형 값을 지정해 원하는 데이터를 반환한다.

4) loc[ ]는 명칭 기반 인덱싱만 가능하다. 따라서 행 위치에 DataFrame 인덱스가 오며, 열 위치에는 칼럼 명을 지정해 원하는 데이터를 반환한다.

5) 명칭 기반 인덱싱에서 슬라이싱을 '시작점:종료점'으로 지정할 때 시작점에서 종료점을 포함한 위치에 있는 데이터를 반환한다.

DataFrame의 ix, iloc, loc는 기본적으로 많이 활용되는 연산자이나, 노련한 개발자들도 혼동하는 부분이 많아서 많은 설명을 하였다.
사용할때 혼동을 줄이기 위해서 일반적으로 loc[ ]나 iloc[ ] 중 하나만 선택해서 사용하는 것이 좋다.

(개인적으로는 주로 loc[ ]만 선택해서 사용하는게 편하다 생각한다.)
