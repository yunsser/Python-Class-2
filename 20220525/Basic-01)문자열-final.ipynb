### % 연산자를 통한 숫자 바로 대입

```python
# 숫자를 바로 대입하는 방법
"현재 온도는 %d도 입니다" % 26 # decimal 십진수 
```

```python
# 문자를 바로 대입하는 방법
"현재 온도는 %s도 입니다" % '이십 육' # s 는 str
```

```python
# 숫자 값을 변수로 대입하여서 적용
num = 3
"I eat %d pythons" % num
```

```python
# 2개 이상의 값을 넣기

num = 10
day = 'five'

"I ate %d pythons so i was sick for %s days" % (num, day)
```

## 포맷코드와 숫자를 통한 정렬 

```python
# 1. 정렬과 공백
"%10s" % 'bigdata' # 오른쪽정렬
"%-10s" % 'bigdata' # 왼쪽정렬
```

```python
# 2. 소수점 표현하기
'%10.2f' % 3.14159212345678
```

## format 함수를 활용한 포매팅

```python
# 숫자 바로 대입하기
"I eat {0} pythons".format(3)
```

```python
# 문자열 바로 대입하기
"I eat {0} pythons".format('five')
```

```python
# 숫자 값을 가진 변수로 대입하기
num1 = 3
"I eat {0} pythons".format(num1)
```

```python
# 숫자 값을 가진 변수로 대입하기
num1 = 3
day1 = 'six'

"I ate {0} pythons so i was sick for {1} days".format(num1, day1)
```

```python
# format 함수를 통한 정렬 
# 왼쪽정렬
'{0:<10}'.format('hi')
# 오른쪽정렬
'{0:>10}'.format('hi')
# 가운데정렬
'{0:^10}'.format('hi')
```

```python
# 왼쪽정렬(공백채우기)
'{0:?<15}'.format('님 나 아심')
# 오른쪽정렬(공백채우기)
'{0:?>10}'.format('hi')
```

```python
## 포맷 함수를 통한 소수점 표현
y = 3.141592123456789

'{0:0.4f}'.format(y) # 원칙 {0:0.4f}
'{0:.4f}'.format(y) # 실무 {0:.4f}

# 오른쪽 정렬을 통한 formatting
'{0:10.4f}'.format(y)
```

## f 문자열 포매팅 

파이썬 3.6버전부터는 f문자열 포매팅 기능을 사용할 수 있다. 
(3.6미만 버전에서는 사용할 수 없는 기능이다.)

```python
name = 'ysp'
age = 20

# format 함수를 통해 활용
'제 이름은 {name}입니다. 나이는 {age}입니다'.format(name='ysp',age=20)

# f 포매팅을 통해 활용
f'제 이름은 {name}입니다. 나이는 {age}입니다'
```

```python
# f 포매팅 변칙
age = 29
f'나는 내년이면 {age+1}살이 됩니다'
```

```python
# f 포매팅을 dict로 표현해보기
dict1 = {'name':'ysp', 'age':20}
```

```python
f'제 이름은 {dict1["name"]}입니다. 나이는 {dict1["age"]}입니다'
```

```python
# 정렬의 방식
f'{"hi":<10}' # 왼쪽정렬
f'{"hi":>10}' # 왼쪽정렬
f'{"hi":^10}' # 왼쪽정렬
```

```python
# 여백을 채우기
# 정렬의 방식
f'{"hi":!<10}' # 왼쪽정렬
f'{"hi":!>10}' # 왼쪽정렬
f'{"hi":!^10}' # 왼쪽정렬
```
